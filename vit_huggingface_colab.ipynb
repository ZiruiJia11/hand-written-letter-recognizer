
# Using a pre-trained Vision Transformer (ViT) with Hugging Face

This notebook demonstrates how to use a pre-trained Vision Transformer (ViT) model from the Hugging Face `transformers` library for image classification.

```python
import torch
from transformers import pipeline

classifier = pipeline(
    task="image-classification",
    model="google/vit-base-patch16-224",
    torch_dtype=torch.float16,
    device=0  # set to 0 or another GPU index if you have a GPU
)
```

```python

from PIL import Image
import requests
from io import BytesIO

# Replace this URL with your own image URL or upload an image to Colab and use its path
image_url = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/pipeline-cat-chonk.jpeg"

# Load the image
response = requests.get(image_url)
img = Image.open(BytesIO(response.content)).convert("RGB")

# Run the classifier on the image
result = classifier(img)
print("Classification result:")
print(result)
```
